```python
import string
import spacy as sp
from spacy.tokenizer import Tokenizer
from spacy.lang.ru import Russian
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import ru_core_news_sm
from collections import Counter
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#1 Южный почтовый Антуан де Сент-Экзюпери
#2 Ночной полет Антуан де Сент-Экзюпери 
#3 Планета людей Антуан де Сент-Экзюпери 
#4 Военный летчик Антуан де Сент-Экзюпери 
#5 Письмо к заложнику Антуан де Сент-Экзюпери 
#6 Маленький принц Антуан де Сент-Экзюпери 

with open('1.txt', 'r', encoding = 'utf-8') as book1:
    text1 = book1.read()
with open('2.txt', 'r', encoding = 'utf-8') as book2:
    text2 = book2.read()
with open('3.txt', 'r', encoding = 'utf-8') as book3:
    text3 = book3.read()
with open('4.txt', 'r', encoding = 'utf-8') as book4:
    text4 = book4.read()
with open('5.txt', 'r', encoding = 'utf-8') as book5:
    text5 = book5.read()
with open('6.txt', 'r', encoding = 'utf-8') as book6:
    text6 = book6.read()
```

```python
nlp = Russian()
nlp.max_length = 10000000
doc = nlp(text1)
words_filtered = nlp(str([token.text for token in doc if not token.is_punct and not token.is_stop]))
words_filtered



word_freq = Counter(token.text for token in words_filtered if token.is_alpha)
most_common_word = str(word_freq.most_common(20))

word_cloud = WordCloud().generate(most_common_word)

plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()

most_common_word

def clean(text1):
    table = str.maketrans("", "", string.punctuation)
    onlyword = text1.translate(table)
    return onlyword

def diversity(text1):
    words = text1.split()
    unique_words = set(words)
    print('        УНИКАЛЬНЫЕ СЛОВА:', len(unique_words))
    print('              ВСЕГО СЛОВ:', len(words))
    return len(unique_words) / len(words)

onlyword = clean(text1)
diversity = diversity(onlyword)

print('ЛЕКСИЧЕСКОЕ РАЗНООБРАЗИЕ:', diversity)
```
