```python
import csv
import re
from googletrans import Translator
from nltk import word_tokenize
import jieba
import pandas as pd 

# 我住在北京。我的父母最近搬到了西安. 现在找房子非常困难。
```

```python
ch_text = input("Введите китайский текст: ")
```

```python
with open('2.7.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    dataset_dict = {}
    for row in reader:
        key = row[0]  
        value = row[1]  
        dataset_dict[key] = value  
print(dataset_dict)
```

```python
translator = Translator()
SENTENSE = ch_text
LANGUAGE_CODE = translator.detect(SENTENSE).lang

k = translator.translate(SENTENSE, dest=LANGUAGE_CODE)
pinyuin_procesed = k.pronunciation.lower()
print(pinyuin_procesed)
```

```python
def replace_tones(text):
    tone_replacements = {
        'ā': 'a1', 'á': 'a2', 'ǎ': 'a3', 'à': 'a4',
        'ē': 'e1', 'é': 'e2', 'ě': 'e3', 'è': 'e4',
        'ī': 'i1', 'í': 'i2', 'ǐ': 'i3', 'ì': 'i4',
        'ō': 'o1', 'ó': 'o2', 'ǒ': 'o3', 'ò': 'o4',
        'ū': 'u1', 'ú': 'u2', 'ǔ': 'u3', 'ù': 'u4',    }

    for vowel, replacement in tone_replacements.items():
        text = re.sub(vowel, replacement, text)

    return text

result = [replace_tones(pinyuin_procesed)]
result = [re.sub(r'[^a-zA-Z0-9а-яА-Я., ]', '', item) for item in result]
print(result) 
```

```python
def replace_text_with_dataset(text, dataset):
    sorted_keys = sorted(dataset.keys(), key=len, reverse=True)
    pattern = re.compile('|'.join(re.escape(key) for key in sorted_keys))
    replaced_text = pattern.sub(lambda match: dataset[match.group(0)], text)
    return replaced_text

result_texts = [replace_text_with_dataset(text, dataset_dict) for text in result]

print(result_texts)    
```

```python
cleaned_text = re.sub(r'[^\u4e00-\u9fff]', '', ch_text)
chinese_text = cleaned_text

cleaned_list = [re.sub(r'[^a-zA-Z0-9а-яА-Я ]', '', item) for item in result_texts]

word_list = jieba.cut(chinese_text, cut_all=False)
#print(list(word_list)) 

ch_text2 = (' '.join(word_list))
rus_text2 = (''.join(cleaned_list))

with open('cyrillic_text.txt', 'w', encoding='utf-8') as file:
        file.write(str(rus_text2))

ch_split = ch_text2.split()
rus_split = rus_text2.split()

df = pd.DataFrame({'Chinese': ch_split, 'Cyrillic': rus_split})

filename = 'table.csv'

df.to_csv(filename, index=False)
df
```
